```python
from datetime import datetime

def getWeekDay(date): 
    unity = date.weekday()
    if (unity == 0):
        return 'Monday'
    elif (unity == 1):
        return 'Tuesday'
    elif (unity == 2):
        return 'Wednesday'
    elif (unity == 3):
        return 'Thursday'
    elif (unity == 4):
        return 'Friday'
    elif (unity == 5):
        return 'Saturday'
    elif (unity == 6):
        return 'Sunday'
    else:
        return 'No Method for this'
def string_to_date_converter(date_string):
    parsed_date = datetime.strptime(date_string, "%Y-%m-%d")
    return parsed_date
```

```python
### Define the functions for the respective actions
from datetime import datetime, timedelta

def check_name(name):
    if name in accepted_names:
        return 'valid'
    elif name in fraud_names:
        return 'fraud'
    else: 
        return 'null'
    
def check_email(email):
    if email in accepted_emails:
        return 'valid'
    elif email in fraud_emails:
        return 'fraud'
    else: 
        return 'null'

def check_transactmode(transmode):
    if transmode in accepted_transactmodes:
        return 'valid'
    elif transmode in fraud_transactmodes:
        return 'fraud'
    else: 
        return 'null'

def check_ips(ip):
    if ip in accepted_ips:
        return 'valid'
    elif ip in fraud_ips:
        return 'fraud'
    else: 
        return 'null'
    
def string_to_date_converter(date_string):
    parsed_date = datetime.strptime(date_string, "%Y-%m-%d")
    return parsed_date

def string_to_time_converter(time_string):
    parsed_time = datetime.strptime(time_string, "%H:%M:%S")
    return parsed_time.hour
```

```python
def set_day_encoder(dsy):
    farsi  = f'Day_{dsy}'
    days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    output_dict = {f'Day_{day}': 0 for day in days_of_week}
    
    if dsy in days_of_week:
        output_dict[farsi] = 1
        
    return output_dict

def set_monthend_category(monsta):
    #MonthEnd_MonthEnd MonthEnd_NormalDay
    farsi = f'MonthEnd_{monsta}'
    monthend_state = ['MonthEnd', 'NormalDay']
    output_dict = {f'MonthEnd_{monthsate}': 0 for monthsate in monthend_state}
    
    if monsta in monthend_state:
        output_dict[farsi] = 1
        
    return output_dict

def is_party_classifier(partycond):
    #'isParty_No', 'isParty_Party'
    farsi = f'isParty_{partycond}'
    party_state = ['No', 'Party']
    output_dict = {f'isParty_{monthsate}': 0 for monthsate in party_state}
    
    if partycond in party_state:
        output_dict[farsi] = 1
        
    return output_dict

def name_classifier(partycond):
    #'checkname_fraud', 'checkname_null','checkname_valid'
    farsi = f'checkname_{partycond}'
    party_state = ['fraud', 'null', 'valid']
    output_dict = {f'checkname_{monthsate}': 0 for monthsate in party_state}
    
    if partycond in party_state:
        output_dict[farsi] = 1
        
    return output_dict

def email_classifier(partycond):
    #'checkmail_fraud', 'checkmail_valid',
    farsi = f'checkmail_{partycond}'
    party_state = ['fraud', 'valid']
    output_dict = {f'checkmail_{monthsate}': 0 for monthsate in party_state}
    
    if partycond in party_state:
        output_dict[farsi] = 1
        
    return output_dict

def tmode_classifier(partycond):
    # 'checktmode_fraud', 'checktmode_valid',
    farsi = f'checktmode_{partycond}'
    party_state = ['fraud', 'valid']
    output_dict = {f'checktmode_{monthsate}': 0 for monthsate in party_state}
    
    if partycond in party_state:
        output_dict[farsi] = 1
        
    return output_dict

def ip_classifier(partycond):
    # 'checkip_fraud','checkip_valid',
    farsi = f'checkip_{partycond}'
    party_state = ['fraud', 'valid']
    output_dict = {f'checkip_{monthsate}': 0 for monthsate in party_state}
    
    if partycond in party_state:
        output_dict[farsi] = 1
        
    return output_dict

def get_time_period(nums):
    #['Night', 'Morning', 'Afternoon', 'Evening'][0, 6, 12, 18, 24
    if (nums >= 0 and nums <6):
        return 'Night'
    elif (nums >= 6 and nums <12):
        return 'Morning'
    elif (nums >= 12 and nums <18):
        return 'Afternoon'
    elif (nums >= 18 and nums <24):
        return 'Evening'
    else:
        return 'No time'

def ctime_classifier(migos):
    partycond = get_time_period(migos)
    #  'ctime_Night', 'ctime_Morning', 'ctime_Afternoon','ctime_Evening'
    farsi = f'ctime_{partycond}'
    party_state = ['Night', 'Morning', 'Afternoon', 'Evening']
    output_dict = {f'ctime_{monthsate}': 0 for monthsate in party_state}
    
    if partycond in party_state:
        output_dict[farsi] = 1
        
    return output_dict
```

```python
#['Amount', 'MonthEnd', 'Day', 'isParty', 'checkname', 'checkmail', 'checktmode', 'checkip', 'cdate', 'ctime', 'transaction_density']

def get_determine(obj_array):
    dubios = []
    name = obj_array['Name']
    #obj_array['checkname'] = check_name(name)
    obj_array.update(name_classifier(check_name(name)))
    
    email = obj_array['Email']
    #obj_array['checkmail'] = check_email(email)
    obj_array.update(email_classifier(check_email(email)))
    
    tranmode = obj_array['TransactionMode']
    #obj_array['checktmode'] = check_transactmode(tranmode)
    obj_array.update(tmode_classifier(check_transactmode(tranmode)))
    
    ipadd = obj_array['IPAddress']
    #obj_array['checkip'] = check_ips(ipadd)
    obj_array.update(ip_classifier(check_ips(ipadd)))
    
    times = obj_array['Time']
    #obj_array['ctime'] = string_to_time_converter(times)
    obj_array.update(ctime_classifier(string_to_time_converter(times)))
    
    date = obj_array['Date']
    obj_array['cdate'] = string_to_date_converter(date)
    
    monthend = obj_array['MonthEnd']
    obj_array.update(set_monthend_category(monthend)) 
    
    dayssd = obj_array['Day']
    obj_array.update(set_day_encoder(dayssd)) 
    
    party = obj_array['isParty']
    obj_array.update(is_party_classifier(party))
    
    
    
    #dubios.append(obj_array)
    
    #data = pd.DataFrame(dubios)
    
    #data['ctime'] = pd.cut(data['ctime'], bins=[0, 6, 12, 18, 24], labels=['Night', 'Morning', 'Afternoon', 'Evening'])
    #data['transaction_density'] = data.groupby('Date')['Amount'].transform('mean')
    
    #data_to_use = data['Amount', 'MonthEnd', 'Day', 'isParty', 'checkname', 'checkmail', 'checktmode', 'checkip', 'ctime', 'transaction_density']
    #data_to_use = data.drop(columns=['Time', 'WeekDay', 'Date', 'isFraud', 'Name','Email', 'IPAddress', 'TransactionMode','MonthEnd', 'Day', 'isParty' ])
    
    return obj_array


def determine_all_array(major):
    dubios = []
    for i in range(len(major)):
        obe_arr = get_determine(major[i])
        dubios.append(obe_arr)
        
        
    data = pd.DataFrame(dubios)
    
    data['transaction_density'] = data.groupby('Date')['Amount'].transform('mean')
    
    data_to_use = data.drop(columns=['Time', 'WeekDay', 'Date', 'isFraud', 'Name','Email', 'IPAddress', 'TransactionMode','MonthEnd', 'Day', 'isParty' ])
    
    return data_to_use
    
```
