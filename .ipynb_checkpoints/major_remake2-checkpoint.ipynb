### Start Creating the Anomaly Detection Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('rematched2.csv')
```

```python
data = data.drop(columns=['Unnamed: 0'])
```

```python
data.head()
```

```python
data['transaction_density'] = data.groupby('Date')['Amount'].transform('mean')
```

```python
data.tail(3)
```

```python
# Converting all features to numerical value
data2 = data.drop(columns=['Time', 'WeekDay', 'isParty', 'isFraud' ])
trans_dense = data['transaction_density']

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ['Name', 'Email', 'TransactionMode','IPAddress', 'MonthEnd', 'Date']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 categorical_features)],
                               remainder='passthrough')

transformed_data = transformer.fit_transform(data2)
transformed_data
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
dat = data.drop(['isFraud'], axis=1)
one_hot_enc = OneHotEncoder()
category_encoded = one_hot_enc.fit_transform(dat[['Name', 'Email', 'IPAddress', 'TransactionMode', 'MonthEnd']]).toarray()
encoded_df = pd.DataFrame(category_encoded, columns=one_hot_enc.get_feature_names_out(['Name', 'Email', 'IPAddress', 'TransactionMode', 'MonthEnd']))
dat = pd.concat([dat, encoded_df], axis=1)
encoded_df.head()
#dat.head()
```

```python
encoded_df['transaction_density'] = data['transaction_density']
encoded_df.tail(50)
```

```python
train_data, test_data = train_test_split(encoded_df, test_size=0.2, random_state=42)
model = IsolationForest(contamination=0.05, random_state=42,)
anomaly_detector = model.fit(train_data)
```

```python
train_data.shape
```

```python
anomaly_prediction = anomaly_detector.predict(test_data)
#anoly_df = pd.DataFrame(anomaly_prediction)
anoly_score = anomaly_detector.decision_function(test_data)
joiner = test_data
joiner['anomaly_analysis'] = anomaly_prediction
joiner['anomaly_score'] = anoly_score
joiner.head(20)
```

```python
import random
def replace_with_anomaly(normal_transact_df, anomaly_transact_df):
    normal_image_df = normal_transact_df
    rows = []
    anomaly_image_df = anomaly_transact_df
    for i in range(len(normal_transact_df)):
        rows.append(i)
        ++i
    fainty = random.sample(rows, len(anomaly_image_df))
    print(fainty)
    counter = 0
    
    for i in fainty:
        name = anomaly_image_df.loc[counter,'Name']
        email = anomaly_image_df.loc[counter,'Email']
        amt = anomaly_image_df.loc[counter,'Amount']
        time = anomaly_image_df.loc[counter,'Time']
        ip_add = anomaly_image_df.loc[counter,'IPAddress']
        transact_mode = anomaly_image_df.loc[counter,'TransactionMode']
        isFraud = anomaly_image_df.loc[counter,'isFraud']
        
        normal_image_df.loc[i, 'Name'] = name
        normal_image_df.loc[i,'Email'] = email
        normal_image_df.loc[i,'Amount'] = amt
        normal_image_df.loc[i,'Time'] = time
        normal_image_df.loc[i,'IPAddress'] = ip_add
        normal_image_df.loc[i,'TransactionMode'] = transact_mode
        normal_image_df.loc[i,'isFraud'] = isFraud
        
        counter = counter + 1
        
    return normal_transact_df
```
