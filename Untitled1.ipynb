```python
import pandas as pd
import random

# Sample Data
makes = ["Toyota", "Nissan", "Volkswagen", "Ford", "Honda"]
colors = ["Red", "Blue", "Yellow", "Green", "Black"]
doors = [3,4,5]
odometer = [random.randint(10000, 90000) for _ in range(100)]
price = [random.randint(10000, 50000) for _ in range(100)]

# Creating the Dataframe
data = {
    "Make": [random.choice(makes) for _ in range(100)],
    "Color": [random.choice(colors) for _ in range(100)],
    "Doors": [random.choice(doors) for _ in range(100)],
    "Odometer(KM)": odometer,
    "Price": price
}

df = pd.DataFrame(data)

print(df.head())
```

```python
df.head(20)
```

```python
average_milage_per_year = 3000
# Categorical Data Encoding
df_encoded = pd.get_dummies(df, columns=["Make", "Color"], drop_first=True)

#Feature Scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_encoded[["Doors", "Odometer(KM)"]] = scaler.fit_transform(df_encoded[["Doors", "Odometer(KM)"]])

# Feature Engineering
df_encoded["Age"] = df_encoded["Odometer(KM)"] / average_milage_per_year

# Target Variable Seperation
X = df_encoded.drop("Price", axis=1)
Y = df_encoded["Price"]

# Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Descricptive Statistics
descriptive_stats = X_train.describe()
descriptive_stats
```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Pairplot to visualize relationships
sns.pairplot(df_encoded, x_vars=["Doors", "Odometer(KM)", "Age"], y_vars="Price", height=5, aspect=0.7, kind="reg")
plt.show()

# Correlation matrix heatmap
correlation_matrix = df_encoded.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
# Outlier Detection
sns.boxplot(x="Make_Toyota", y="Price", data=df_encoded)
plt.show()
```

```python
# Model Training
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, Y_train)
feature_importances = rf.feature_importances_
```
